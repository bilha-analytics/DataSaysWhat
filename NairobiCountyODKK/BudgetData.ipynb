```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.style.use("seaborn-dark")

plt.rcParams['figure.figsize'] = (15.0, 10.0)
```

```python
df_nbo_pa = pd.read_csv("Local_Authority_Expenditures_by_Year_Nairobi.csv")
#df_nbo_pa.head(2)
```

```python
df_nbo_pa = pd.melt( df_nbo_pa, id_vars=["County"], value_vars=df_nbo_pa.columns[2:-1])
#df_nbo_pa
```

```python
print("# Nairobi County Expenditure - 2002 to 2008")
plt.plot(df_nbo_pa.variable, df_nbo_pa.value)
```
